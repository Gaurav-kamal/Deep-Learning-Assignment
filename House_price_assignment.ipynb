#Handling Missing Data

```python
# Replacing "NA" string value to NaN

df1.replace({'NA':np.nan}, inplace = True)
df1.head()
```

```python
# Fill missing LotFrontage values with the median grouped by Neighborhood
df['LotFrontage'] = df.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))
```

```python
df.head()
```

```python
# Fillinf missing values with 0 and 'none'

df['GarageYrBlt'].fillna(0, inplace=True)
df['GarageType'].fillna('None', inplace=True)
df['GarageFinish'].fillna('None', inplace=True)
```

```python
print(df[['GarageYrBlt', 'GarageType', 'GarageFinish']].isnull().sum())
```

```python
# Filling missing values by creating new category

df['Alley'].fillna('No Alley', inplace=True)
df['PoolQC'].fillna('No Pool', inplace=True)
df['Fence'].fillna('No Fence', inplace=True)
```

```python
print(df[['Alley', 'PoolQC', 'Fence']].isnull().sum())
```

#Feature Creation

```python
# Create a new feature for TotalSF (Total Square Footage)
df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
```

```python
# Create Age of House feature
df['Age_of_House'] = df['YrSold'] - df['YearBuilt']
```

```python
# Create a binary feature for Has Pool
df['Has_Pool'] = df['PoolArea'].apply(lambda x: 1 if x > 0 else 0)
```

```python
print("New features (TotalSF, Age_of_House, Has_Pool):")
print(df[['TotalSF', 'Age_of_House', 'Has_Pool']].head())
```

#Transforming Features

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Encode categorical features using One-Hot Encoding
df = pd.get_dummies(df, columns=['Neighborhood', 'GarageType', 'Alley'], drop_first=True)

# Log transformation for skewed numerical features
df['SalePrice'] = np.log1p(df['SalePrice'])
df['GrLivArea'] = np.log1p(df['GrLivArea'])
df['TotalSF'] = np.log1p(df['TotalSF'])
```
